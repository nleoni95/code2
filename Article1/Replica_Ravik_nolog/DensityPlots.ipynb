```python
#on regarde les samples bayes et FMP. kom je pri. ah j'ai pas Ã©crit les samples bayes lul.
########################### on passe aux densitÃ©s calibration multiples
colorfz='cyan'
colorf='tab:blue'
colorfzkoh='gold'
colorfkoh='tab:orange'
colorfzfb='springgreen'
colorffb='tab:green'
colorpri='tab:red'
lw2=2
lw1=2
foldname='results/'
foldnamefmp='results/fmp/'
foldnamefb='results/bayes/'
foldnamekoh='results/koh/'
###load dfs
sampfmp=pd.DataFrame(data=np.loadtxt(foldnamefmp+'sample.gnu')[:,:3],columns=[r'$\theta_1$',r'$\theta_2$',r'$\theta_3$'])
sampfb=pd.DataFrame(data=np.loadtxt(foldnamefb+'sample2.gnu')[:,:3],columns=[r'$\theta_1$',r'$\theta_2$',r'$\theta_3$'])
sampkoh=pd.DataFrame(data=np.loadtxt(foldnamekoh+'sample.gnu')[:,:3],columns=[r'$\theta_1$',r'$\theta_2$',r'$\theta_3$'])
samppri=pd.DataFrame(data=np.loadtxt('results/priorsamples.gnu')[-len(sampfmp):,:3],columns=[r'$\theta_1$',r'$\theta_2$',r'$\theta_3$'])



sampfmp['Calibration']='FMP'
sampfb['Calibration']='Full Bayes'
sampkoh['Calibration']='KOH'
samppri['Calibration']='Prior'
df=pd.concat([sampfmp,sampfb,sampkoh,samppri],ignore_index=True)
# convertir les colonnes des paramÃ¨tres dans l'espace rÃ©el. valeur ominales : 40, 18.9e-6, 0.75.
df[r'$\theta_1$']=(df[r'$\theta_1$']+0.5)*40
df[r'$\theta_2$']=(df[r'$\theta_2$']+0.5)*18.9E-6
df[r'$\theta_3$']=(df[r'$\theta_3$']+0.5)*0.75

mpl.rcParams["axes.labelsize"]=18
mpl.rcParams['figure.figsize'] = 6,6
ax=sns.kdeplot(x=r'$\theta_3$',data=df,hue='Calibration',palette={
                    'FMP':colorf,
                    'KOH':colorfkoh,
                    'Full Bayes':colorffb,
                    'Prior':colorpri
                })
plt.setp(ax.get_legend().get_texts(), fontsize='16') # for legend text
plt.setp(ax.get_legend().get_title(), fontsize='18') # for legend title
plt.tick_params(axis='both', labelsize=15)

g = sns.PairGrid(df,hue='Calibration',diag_sharey=False,
                palette={
                    'FMP':colorf,
                    'KOH':colorfkoh,
                    'Full Bayes':colorffb
                    ,'Prior':colorpri
                },
                 height=3,aspect=1
                )
g.map_diag(sns.kdeplot)
axes=g.axes
g.map_upper(sns.kdeplot,levels=2)
g.map_lower(sns.scatterplot,s=1)
g.add_legend()

```

```python

#prÃ©dictions avec densitÃ©s multiples.
colorfz='cyan'
colorf='tab:blue'
colorfzkoh='gold'
colorfkoh='tab:orange'
colorfzfb='springgreen'
colorffb='tab:green'
colorpri='tab:grey'
lw2=2
lw1=2
case=16
#le cas 8 a des observations relou.
foldname='results/'
obs=pd.DataFrame(data=np.loadtxt(foldname+'/obs/obs'+str(case)+'.gnu'),columns=['x','y','incx','incy'])

pfb=pd.DataFrame(data=np.loadtxt(foldname+'bayes/preds'+str(case)+'.gnu'),columns=['x','fzmean','fzstd','fzstd2','fmean','q2p5','q97p5'])
sampFfb=pd.DataFrame(data=np.loadtxt(foldname+'bayes/sampsf'+str(case)+'.gnu'),columns=['x','s1','s2','s3','s4','s5','s6','s7','s8','s9','s10'])
sampZfb=pd.DataFrame(data=np.loadtxt(foldname+'bayes/sampsz'+str(case)+'.gnu'),columns=['x','s1','s2','s3','s4','s5','s6','s7','s8','s9','s10'])

pfmp=pd.DataFrame(data=np.loadtxt(foldname+'fmp/preds'+str(case)+'.gnu'),columns=['x','fzmean','fzstd','fzstd2','fmean','q2p5','q97p5'])
sampFfmp=pd.DataFrame(data=np.loadtxt(foldname+'fmp/sampsf'+str(case)+'.gnu'),columns=['x','s1','s2','s3','s4','s5','s6','s7','s8','s9','s10'])
sampZfmp=pd.DataFrame(data=np.loadtxt(foldname+'fmp/sampsz'+str(case)+'.gnu'),columns=['x','s1','s2','s3','s4','s5','s6','s7','s8','s9','s10'])

pkoh=pd.DataFrame(data=np.loadtxt(foldname+'koh/preds'+str(case)+'.gnu'),columns=['x','fzmean','fzstd','fzstd2','fmean','q2p5','q97p5'])
sampFkoh=pd.DataFrame(data=np.loadtxt(foldname+'koh/sampsf'+str(case)+'.gnu'),columns=['x','s1','s2','s3','s4','s5','s6','s7','s8','s9','s10'])
sampZkoh=pd.DataFrame(data=np.loadtxt(foldname+'koh/sampsz'+str(case)+'.gnu'),columns=['x','s1','s2','s3','s4','s5','s6','s7','s8','s9','s10'])


fig,axs=plt.subplots(nrows=1,ncols=4,figsize=(20,6))



axs=np.reshape(axs,(2,2))
axs[0,0].plot(pfb['x'],sampFfb['s1'],linewidth=1.5,color=colorfzfb,zorder=0,alpha=0.8,label=r'samples of $f+z_{\theta}$')
axs[0,0].plot(pfb['x'],pfb['fmean'],linewidth=lw2,linestyle='-.',color=colorf,label=r'$f(x,\theta_{\mathrm{pmean}})$')
axs[0,0].plot(pfb['x'],pfb['q97p5'],linewidth=lw1,linestyle='--',color=colorf,label=r'95$\%$ credence interval')
axs[0,0].plot(pfb['x'],pfb['q2p5'],linewidth=lw1,linestyle='--',color=colorf)

axs[0,1].plot(pfmp['x'],sampFfmp['s1'],linewidth=1.5,color=colorfz,zorder=0,alpha=0.8,label=r'samples of $f+z_{\theta}$')
axs[0,1].plot(pfmp['x'],pfmp['fmean'],linewidth=lw2,linestyle='-.',color=colorf,label=r'$f(x,\theta_{\mathrm{pmean}})$')
axs[0,1].plot(pfmp['x'],pfmp['q97p5'],linewidth=lw1,linestyle='--',color=colorf,label=r'95$\%$ credence interval')
axs[0,1].plot(pfmp['x'],pfmp['q2p5'],linewidth=lw1,linestyle='--',color=colorf)

axs[1,0].plot(pkoh['x'],sampFkoh['s1'],linewidth=1.5,color=colorfzkoh,zorder=0,alpha=0.8,label=r'samples of $f+z_{\theta}$')
axs[1,0].plot(pkoh['x'],pkoh['fmean'],linewidth=lw2,linestyle='-.',color=colorf,label=r'$f(x,\theta_{\mathrm{pmean}})$')
axs[1,0].plot(pkoh['x'],pkoh['q97p5'],linewidth=lw1,linestyle='--',color=colorf,label=r'95$\%$ credence interval')
axs[1,0].plot(pkoh['x'],pkoh['q2p5'],linewidth=lw1,linestyle='--',color=colorf)


ls=['s1','s2','s3','s4','s5','s6','s7','s8','s9']
#ls=['s5']

for s in ls:
    axs[0,0].plot(pfb['x'],sampFfb[s]+sampZfb[s],linewidth=1.5,color=colorfz,zorder=0,alpha=0.8)
    axs[0,1].plot(pfmp['x'],sampFfmp[s]+sampZfmp[s],linewidth=1.5,color=colorfz,zorder=0,alpha=0.8)
    axs[1,0].plot(pkoh['x'],sampFkoh[s]+sampZkoh[s],linewidth=1.5,color=colorfz,zorder=0,alpha=0.8)

axs[0,0].errorbar(obs['x'],obs['y'],xerr=3*obs['incx'],
             color='black',marker='o',ms=6,mfc='white',
             capsize=5,linewidth=0,elinewidth=2,
             label='observations',zorder=100)

axs[0,1].errorbar(obs['x'],obs['y'],xerr=3*obs['incx'],
             color='black',marker='o',ms=6,mfc='white',
             capsize=5,linewidth=0,elinewidth=2,
             label='observations',zorder=100)
axs[1,1].errorbar(obs['x'],obs['y'],xerr=3*obs['incx'],
             color='black',marker='o',ms=6,mfc='white',
             capsize=5,linewidth=0,elinewidth=2,
             label='observations',zorder=100)

axs[1,0].errorbar(obs['x'],obs['y'],xerr=3*obs['incx'],
             color='black',marker='o',ms=6,mfc='white',
             capsize=5,linewidth=0,elinewidth=2,
             label='observations',zorder=100)

for ax in axs:
    for ax2 in ax:
        ax2.yaxis.offsetText.set_visible(False)

offset = axs[0,0].yaxis.get_major_formatter().get_offset()

axs[0,0].set_xlim([0,35])
axs[0,0].set_ylim([0,obs['y'][len(obs['y'])-1]+3*obs['incy'][len(obs['y'])-1]+1e5])
axs[0,0].set_xlabel('Wall superheat $[K]$ ')
axs[0,0].yaxis.set_label_text(r'Heat flux $[W.m^{-2}] \times 10^6$')
axs[0,0].set_title('FB predictions, case '+str(case),fontsize=25)

axs[0,1].set_xlim([0,35])
axs[0,1].set_ylim([0,obs['y'][len(obs['y'])-1]+3*obs['incy'][len(obs['y'])-1]+1e5])
axs[0,1].set_xlabel('Wall superheat $[K]$')
#axs[1].set_ylabel('Heat flux $[W.m^{-2}]$')
axs[0,1].set_title('FMP predictions, case '+str(case),fontsize=25)

axs[1,1].set_xlim([0,35])
axs[1,1].set_ylim([0,obs['y'][len(obs['y'])-1]+3*obs['incy'][len(obs['y'])-1]+1e5])
axs[1,1].set_xlabel('Wall superheat $[K]$')
#axs[1].set_ylabel('Heat flux $[W.m^{-2}]$')
axs[1,0].set_title('KOH predictions, case '+str(case),fontsize=25)

axs[1,0].set_xlim([0,35])
axs[1,0].set_ylim([0,obs['y'][len(obs['y'])-1]+3*obs['incy'][len(obs['y'])-1]+1e5])
axs[1,0].set_xlabel('Wall superheat $[K]$')
#axs[1].set_ylabel('Heat flux $[W.m^{-2}]$')
axs[1,1].set_title('KOH pooled predictions, case '+str(case),fontsize=25)


#print(obs)
plt.tight_layout()

axs[0,0].legend(ncol=5,bbox_to_anchor=(0.8, 0.03),
          bbox_transform=fig.transFigure)
#print les hpars koh pooled
#ls=['s5']
#cases=['4','5','6','15','18','20','21','22']# ceux qui sont bons.
```
