```python
#affichage de statistiques sur les r√©sultats de calcul.
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#affichage slices
foldname=''
slices=pd.DataFrame(data=np.loadtxt(foldname+'results/slice.gnu'),columns=['t1','t2','t3','t4','t5','h1','h2','h3','h4','h5','h6','score'])
sliceas50=pd.DataFrame(data=np.loadtxt(foldname+'results/as/slice50.gnu'),columns=['t1','t2','t3','t4','t5','h1','h2','h3','h4','h5','h6','score'])
sliceas150=pd.DataFrame(data=np.loadtxt(foldname+'results/as/slice150.gnu'),columns=['t1','t2','t3','t4','t5','h1','h2','h3','h4','h5','h6','score'])
sliceas250=pd.DataFrame(data=np.loadtxt(foldname+'results/as/slice250.gnu'),columns=['t1','t2','t3','t4','t5','h1','h2','h3','h4','h5','h6','score'])
sliceas350=pd.DataFrame(data=np.loadtxt(foldname+'results/as/slice350.gnu'),columns=['t1','t2','t3','t4','t5','h1','h2','h3','h4','h5','h6','score'])


slices['scoreexp']=np.exp(slices['score'])
sliceas50['scoreexp']=np.exp(sliceas50['score'])
sliceas150['scoreexp']=np.exp(sliceas150['score'])
sliceas250['scoreexp']=np.exp(sliceas250['score'])
sliceas350['scoreexp']=np.exp(sliceas350['score'])

figsize=(8,6)
lw=2
#affichage score
fig,ax=plt.subplots(ncols=1,nrows=1,figsize=figsize)
ax.scatter(slices['t1'],slices['scoreexp'],color='black',label="Reference")
ax.plot(sliceas50['t1'],sliceas50['scoreexp'],label='AS50',color='tab:red',linewidth=lw)
ax.plot(sliceas150['t1'],sliceas150['scoreexp'],label='AS150',color='tab:red',linewidth=lw,linestyle=':')
ax.plot(sliceas250['t1'],sliceas250['scoreexp'],label='AS250',color='tab:red',linewidth=lw,linestyle='--')
ax.plot(sliceas350['t1'],sliceas350['scoreexp'],label='AS350',color='tab:red',linewidth=lw,linestyle='-.')
ax.set_xlabel(r'$\theta_1$',fontsize=20)
ax.set_ylabel(r'$\mathrm{p}(\mathbf{y}_{\mathrm{obs}}|\mathbf{\theta},\widehat{\mathbf{\psi}}_{\mathrm{FMP}}(\mathbf{\theta}))$',fontsize=20)
ax.tick_params(axis='both', labelsize=16)
ax.yaxis.get_offset_text().set_fontsize(16)
ax.legend(fontsize=16,loc='best',frameon=False)
ax.set_xlim([0,1])
#ax.set_ylim([0,1.7e20])


#affichage sigma
crit='h1'
fig,ax=plt.subplots(ncols=1,nrows=1,figsize=figsize)
ax.scatter(slices['t1'],slices[crit],color='black',label="Reference")
ax.plot(sliceas50['t1'],sliceas50[crit],label='AS50',color='tab:red',linewidth=lw)
ax.plot(sliceas150['t1'],sliceas150[crit],label='AS150',color='tab:red',linewidth=lw,linestyle=':')
ax.plot(sliceas250['t1'],sliceas250[crit],label='AS250',color='tab:red',linewidth=lw,linestyle='--')
ax.plot(sliceas350['t1'],sliceas350[crit],label='AS350',color='tab:red',linewidth=lw,linestyle='-.')
ax.tick_params(axis='both', labelsize=16)
ax.legend(fontsize=16,loc='best',frameon=False)
ax.set_xlim([0,1])
ax.set_xlabel(r'$\theta_1$',fontsize=20)
ax.set_ylabel(r'$\widehat{\log \sigma}_{\mathrm{FMP}}$',fontsize=20)
ax.tick_params(axis='both', labelsize=16)
```

```python
#affichage de samples de f+z.

mpl.rcParams['font.size']=16
foldname=''
oa=pd.DataFrame(data=np.loadtxt(foldname+"results/obsalpha.gnu"),columns=['x','alpha'])
od=pd.DataFrame(data=np.loadtxt(foldname+"results/obsdiam.gnu"),columns=['x','diam'])

fa=pd.DataFrame(data=np.transpose(np.loadtxt(foldname+'results/as/sampsF_alpha.gnu')),columns=['s1','s2','s3','s4','s5','s6','s7','s8','s9','s10'])
fd=pd.DataFrame(data=np.transpose(np.loadtxt(foldname+'results/as/sampsF_diam.gnu')),columns=['s1','s2','s3','s4','s5','s6','s7','s8','s9','s10'])
za=pd.DataFrame(data=np.transpose(np.loadtxt(foldname+'results/as/sampsZ_alpha.gnu')),columns=['s1','s2','s3','s4','s5','s6','s7','s8','s9','s10'])
zd=pd.DataFrame(data=np.transpose(np.loadtxt(foldname+'results/as/sampsZ_diam.gnu')),columns=['s1','s2','s3','s4','s5','s6','s7','s8','s9','s10'])

#alpha
atransp=0.7
ms=20
figsize=(6,5)
fig,ax=plt.subplots(figsize=figsize)

plt.plot(oa['x'],fa['s1']+za['s1'],label='posterior samples',color='grey',alpha=atransp)
for l in fa.columns:
    plt.plot(oa['x'],fa[l]+za[l],color='grey',alpha=atransp)
ax.scatter(oa['x'],oa['alpha'],color='red',s=ms,marker='o',label='observations')

#ax.set_ylim([0,0.4])
ax.set_xlim([0,oa['x'][len(oa['x'])-1]])
ax.tick_params(axis='both', labelsize=16)
ax.ticklabel_format(axis='x',style='sci',scilimits=(0,0))
ax.set_ylabel(r'$\alpha\,[-]$',fontsize=20)
ax.set_xlabel(r'$r\,[m]$',fontsize=20)
ax.legend(fontsize=16,loc="upper left") 

#diam√®tre
fig,ax=plt.subplots(figsize=figsize)

plt.plot(od['x'],fd['s1']+zd['s1'],label='posterior samples',color='grey',alpha=atransp)
for l in fa.columns:
    plt.plot(od['x'],fd[l]+zd[l],color='grey',alpha=atransp)
ax.scatter(od['x'],od['diam'],color='red',s=ms,marker='o',label='observations')

#ax.set_ylim([0,0.4])
ax.set_xlim([0,od['x'][len(od['x'])-1]])
ax.tick_params(axis='both', labelsize=16)
ax.ticklabel_format(axis='both',style='sci',scilimits=(0,0))
ax.set_ylabel(r'$D\,[m]$',fontsize=20)
ax.set_xlabel(r'$r\,[m]$',fontsize=20)
ax.legend(fontsize=16,loc="lower left") 

```

```python
# Erreur du surrogate DEBORA
# Void fraction
surpres_alpha=pd.DataFrame(data=np.loadtxt("results/errz_alpha.gnu"),columns=['nmodes','projerr','gperr','errtot'])
df2=surpres_alpha

fig,(ax1)=plt.subplots(figsize=(6,5))
ax1.fill_between(df2['nmodes'],0,df2['projerr']*df2['errtot'],facecolor='none',edgecolor='black',hatch='//',label='projection error')
ax1.fill_between(df2['nmodes'],df2['projerr']*df2['errtot'],df2['errtot'],alpha=0.3,color='black',label="learning error")
ax1.legend(fontsize=20)
ax1.set_xlim([1,8])
ax1.set_ylim([0,8])

ax1.vlines(4,0,3,color='tab:red',linewidth=4)
ax1.set_xlabel(r'$\#$ of principal modes',fontsize=20)
ax1.set_ylabel(r'relative $L_2$ error [$\%$]',fontsize=20)
ax1.set_title('Void fraction surrogate',fontsize=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)

#Diam√®tre
surpres_diam=pd.DataFrame(data=np.loadtxt("results/errz_diam.gnu"),columns=['nmodes','projerr','gperr','errtot'])
df2=surpres_diam

fig,(ax1)=plt.subplots(figsize=(6,5))
ax1.fill_between(df2['nmodes'],0,df2['projerr']*df2['errtot'],facecolor='none',edgecolor='black',hatch='//',label='projection error')
ax1.fill_between(df2['nmodes'],df2['projerr']*df2['errtot'],df2['errtot'],alpha=0.3,color='black',label="learning error")
ax1.legend(fontsize=20)
ax1.set_xlim([1,8])
ax1.set_ylim([0,13])

ax1.vlines(6,0,3,color='tab:red',linewidth=4)
ax1.set_xlabel(r'$\#$ of principal modes',fontsize=20)
ax1.set_ylabel(r'relative $L_2$ error [$\%$]',fontsize=20)
ax1.set_title('Bubble diameter surrogate',fontsize=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
```

```python
# affichage des moyennes / variances de chaque post√©rieure d'apr√®s les samples.
foldname='results_serveur/'
dfas50=pd.DataFrame(data=np.loadtxt(foldname+'as/samp_alpha_50.gnu')[:,:5],columns=['t1','t2','t3','t4','t5'])
dfas150=pd.DataFrame(data=np.loadtxt(foldname+'as/samp_alpha_150.gnu')[:,:5],columns=['t1','t2','t3','t4','t5'])
dfas250=pd.DataFrame(data=np.loadtxt(foldname+'as/samp_alpha_250.gnu')[:,:5],columns=['t1','t2','t3','t4','t5'])
dfas350=pd.DataFrame(data=np.loadtxt(foldname+'as/samp_alpha_350.gnu')[:,:5],columns=['t1','t2','t3','t4','t5'])
dfas450=pd.DataFrame(data=np.loadtxt(foldname+'as/samp_alpha_450.gnu')[:,:5],columns=['t1','t2','t3','t4','t5'])
dfas550=pd.DataFrame(data=np.loadtxt(foldname+'as/samp_alpha_550.gnu')[:,:5],columns=['t1','t2','t3','t4','t5'])

dffmp=pd.DataFrame(data=np.loadtxt(foldname+'fmp/samp_alpha.gnu')[:,:5],columns=['t1','t2','t3','t4','t5'])

def print_table(crit):
    a1=np.array(dffmp.describe().loc[['mean','std'],crit])
    a2=np.array(dfas50.describe().loc[['mean','std'],crit])
    a3=np.array(dfas150.describe().loc[['mean','std'],crit])
    a4=np.array(dfas250.describe().loc[['mean','std'],crit])
    print("$%.2f (%.2f)$& $%.2f (%.2f)$&$%.2f (%.2f)$&$%.2f (%.2f)$"%(a1[0],a1[1],a2[0],a2[1],a3[0],a3[1],a4[0],a4[1]))
    
print(dffmp.describe())
print(dfas550.describe())
```
