```python
#affichage de statistiques sur les résultats de calcul.
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def plot_calcul(nr):
    filename='../data/qmc/'+str(nr)+'/clean_profile.dat'
    df=pd.DataFrame(data=np.loadtxt(filename,skiprows=1),columns=['pos','alpha','diam','V1','V2'])
    exp=np.loadtxt("../data/exp/A6.dat")
    #on garde seulement la deuxième colonne 1 (-R) et (colonne 11 alpha) (column 14 diam) (column 12 V)
    expo=exp[:-4,[1,11,14,12]]
    expo[:,0]*=-1
    fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize=(18,6))
    ax1.scatter(df['pos'],df['alpha'],color='tab:blue',label='calcul')
    ax1.scatter(expo[:,0],expo[:,1],color='tab:red',label='mesures')
    ax1.set_title('taux de vide')
    ax1.legend()
    
    ax2.scatter(df['pos'],df['diam'],color='tab:blue')
    ax2.scatter(expo[:,0],expo[:,2],color='tab:red')
    ax2.set_title('diamètre')
    
    ax3.scatter(df['pos'],df['V1'],color='tab:blue',label='V1')
    ax3.scatter(df['pos'],df['V2'],color='tab:green',label='V2')
    ax3.scatter(expo[:,0],expo[:,3],color='tab:red')
    ax3.set_title('vitesse')
    
    plt.tight_layout()
    plt.show()
    
```

```python
stats=pd.DataFrame(data=np.loadtxt("results/stats.gnu"),columns=['tauxparoi','tauxcoeur','tauxmoyen','diamparoi','diamcoeur','calcnr','l2diamfull','l2diampartial','l2V1','l2V2','phi','BK','COAL','NUCL','MT'])
tdvgphi=pd.DataFrame(data=np.loadtxt("results/tdvmoy0.gnu"),columns=['phi','tdv'])
tdvgbk=pd.DataFrame(data=np.loadtxt("results/tdvmoy1.gnu"),columns=['bk','tdv'])
tdvgcoal=pd.DataFrame(data=np.loadtxt("results/tdvmoy2.gnu"),columns=['coal','tdv'])
tdvgnucl=pd.DataFrame(data=np.loadtxt("results/tdvmoy3.gnu"),columns=['nucl','tdv'])
tdvgmt=pd.DataFrame(data=np.loadtxt("results/tdvmoy4.gnu"),columns=['mt','tdv'])

oatphi=pd.DataFrame(data=np.loadtxt("results/oat0.gnu"),columns=['phi','tvmean','tvstd']) #3 écarts-type.
oatbk=pd.DataFrame(data=np.loadtxt("results/oat1.gnu"),columns=['bk','tvmean','tvstd']) #3 écarts-type.
oatcoal=pd.DataFrame(data=np.loadtxt("results/oat2.gnu"),columns=['coal','tvmean','tvstd']) #3 écarts-type.
oatnucl=pd.DataFrame(data=np.loadtxt("results/oat3.gnu"),columns=['nucl','tvmean','tvstd']) #3 écarts-type.
oatmt=pd.DataFrame(data=np.loadtxt("results/oat4.gnu"),columns=['mt','tvmean','tvstd']) #3 écarts-type.


```

```python
mpl.rcParams['text.usetex']=True
mpl.rcParams['font.sans-serif']='Arial'
mpl.rcParams['font.weight']='bold'
#mpl.rcParams['axes.labelsize']= 33.0 #def value : 11.0
#mpl.rcParams['axes.titlesize']= 28.0 #def value : 12.0
mpl.rcParams['savefig.format']='png'
mpl.rcParams['font.size']=22
mpl.rcParams['legend.fontsize']=18
```

```python
fig, ((ax1,ax2,ax3),(ax4,ax5,ax6)) = plt.subplots(2,3, figsize=(18,10))

#vraies valeurs expérimentales
truetp=0.3116
truetc=0.0656
truetm=0.208
truedp=0.000176
truedc=0.000447
nbins=30
lw=3

ax1.hist(stats['tauxparoi'],bins=nbins,density=True,alpha=0.8)
ax1.set_title('Taux de vide à la paroi')
ax1.vlines(truetp,0,5,color='black',lw=lw)
ax1.axes.yaxis.set_visible(False)

ax2.hist(stats['tauxcoeur'],bins=nbins,density=True,alpha=0.8)
ax2.set_title('Taux de vide à coeur')
ax2.vlines(truetc,0,10,color='black',lw=lw)
ax2.axes.yaxis.set_visible(False)

ax3.hist(stats['tauxmoyen'],bins=nbins,density=True,alpha=0.8)
ax3.set_title('Taux de vide moyen')
ax3.vlines(truetm,0,5,color='black',lw=lw)
ax3.axes.yaxis.set_visible(False)

ax4.hist(stats['diamparoi'],bins=nbins,density=True,alpha=0.8)
ax4.set_title('Diamètre à la paroi')
ax4.vlines(truedp,0,8000,color='black',lw=lw)
ax4.axes.yaxis.set_visible(False)

ax5.hist(stats['diamcoeur'],bins=nbins,density=True,alpha=0.8)
ax5.set_title('Diamètre à coeur')
ax5.vlines(truedc,0,7000,color='black',lw=lw)
ax5.axes.yaxis.set_visible(False)

ax6.hist(stats['l2V1'],bins=nbins,density=True,alpha=0.8)
ax6.set_title('Erreur sur le profil de vitesse')
ax6.vlines(0,0,5,color='black',lw=lw)
ax6.axes.yaxis.set_visible(False)

plt.tight_layout()
plt.show()

```

```python
#recherche des 'bons' calculs
#puisque la bonne prédiction du diamètre à coeur est très rare, on trie selon ce critère.
print(stats.query('diamcoeur<0.00047').query('diamcoeur>0.00041').query('diamparoi>0.0001').query('diamparoi<0.0002').query('tauxparoi<0.5'))
#ccl : les calculs qui prédisent le diam à coeur correctement surprédisent grandement le taux de vide (x2 !)
#c'est toujours avec BK + et COAL +.
#on n'arrive pas vraiment à trouver une bonne courbe globale pour le diamètre.
```

```python
#calcul du taux de vide moyen des données expérimentales
exp=np.loadtxt("../data/exp/A6.dat")
#on garde seulement la deuxième colonne 1 (-R) et la douzième colonne 11 alpha 
expo=exp[:-4,[1,11]]
#multiplication par -1 de la position
expo[:,0]*=-1
tm=0
for i in range(0,expo.shape[0]-1):
    tm+=0.5*(expo[i,1]+expo[i+1,1])*(expo[i,0]**2-expo[i+1,0]**2)
tm/=expo[0,0]**2
print(tm)
```
