```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
#inline backend pour voir les figures dans le frontend notebook https://ipython.readthedocs.io/en/stable/interactive/plotting.html

observations=pd.DataFrame(data=np.loadtxt('results/observations.gnu'),columns=['x','obs']) #dataframe qui contient les observations sur lesquelles on travaille

post1=pd.DataFrame(data=np.loadtxt('results/logp.gnu'),columns=['t1','t2','evidence']) #DataFrame qui contient la postérieure des hyperparamètres, les meilleures valeurs associées à chaque couple de paramètres.
post2=pd.DataFrame(data=np.loadtxt('results/logp_koh.gnu'),columns=['t1','t2','evidence_koh']) #DataFrame qui contient la postérieure des hyperparamètres, les meilleures valeurs associées à chaque couple de
post3=pd.DataFrame(data=np.loadtxt('results/logp_classical.gnu'),columns=['t1','t2','evidence_classical']) #DataFrame qui contient la postérieure des hyperparamètres, les meilleures valeurs associées à chaque couple de

gp1=pd.DataFrame(data=np.loadtxt('results/gpopti.gnu'),columns=['t1','t2','evidence']) #DataFrame qui contient la postérieure des hyperparamètres, les meilleures valeurs associées à chaque couple de paramètres.
gp2=pd.DataFrame(data=np.loadtxt('results/gpkoh.gnu'),columns=['t1','t2','evidence_koh']) #DataFrame qui contient la postérieure des hyperparamètres, les meilleures valeurs associées à chaque couple de
gp3=pd.DataFrame(data=np.loadtxt('results/gpclassical.gnu'),columns=['t1','t2','evidence_classical']) #DataFrame qui contient la postérieure des hyperparamètres, les meilleures valeurs associées à chaque couple de

samples_opti=pd.DataFrame(data=np.loadtxt('results/samples_opti.gnu'),columns=['t1','t2'])#Evaluation des logpostérieures. col1 : theta1, col2: theta2, col3: true logpost, col4 : logpost gp hpars, col5: gp logp
samples_koh=pd.DataFrame(data=np.loadtxt('results/samples_koh.gnu'),columns=['t1','t2'])#Evaluation des logpostérieures. col1 : theta1, col2: theta2, col3: true logpost, col4 : logpost gp hpars, col5: gp logp

hpars=pd.DataFrame(data=np.loadtxt('results/hpars_opti.gnu'),columns=['t1','t2','edm','exp','lcor'])

pred=pd.DataFrame(data=np.loadtxt('results/pred_opti.gnu'),columns=['x','fmean','flow','fsup','zmean','zlow','zsup','f+zmean','f+zlow','f+zsup']) #DataFramequi

pred_classical=pd.DataFrame(data=np.loadtxt('results/pred_classical.gnu'),columns=['x','fmean','flow','fsup','zmean','zlow','zsup','f+zmean','f+zlow','f+zsup']) #DataFrame

pred_koh=pd.DataFrame(data=np.loadtxt('results/pred_koh.gnu'),columns=['x','fmean','flow','fsup','zmean','zlow','zsup','f+zmean','f+zlow','f+zsup']) #DataFrame
```

```python
post_params=pd.merge(post1,post2)
post_params=pd.merge(post_params,post3)

post_gp=pd.merge(gp1,gp2)
post_gp=pd.merge(post_gp,gp3)

#définition de la vérité


#paramètres de sauvegarde
save=True
file_name='results/20obs/20obs'
nobs=observations.shape[0]
```

```python
#Définition du style de plots
plt.style.use('seaborn-white')
# pour la customisation des rcParams : https://matplotlib.org/3.2.0/tutorials/introductory/customizing.html
#print(mpl.rcParams)
mpl.rcParams['axes.labelsize']= 20.0 #def value : 11.0
mpl.rcParams['axes.titlesize']= 20.0 #def value : 12.0
mpl.rcParams['font.size']= 20.0 #def value : 10.0
mpl.rcParams['lines.markersize']= 15.0 #def value : 7.0
mpl.rcParams['lines.linewidth']= 3 #def value : 1.75
mpl.rcParams['legend.fontsize']= 18.0 #def value : 10.0
mpl.rcParams['legend.frameon']= True
mpl.rcParams['text.usetex']= False
mpl.rcParams['xtick.labelsize']= 15.0 #def value : 10.0
mpl.rcParams['xtick.major.size']= 10.0 #taille des ticks
mpl.rcParams['ytick.labelsize']= 15.0 #def value : 10.0
mpl.rcParams['ytick.major.size']= 10.0 #taille des ticks
```

```python
#Contourplot de la postérieure. On peut avoir des couleurs continues avec imshow, voir https://jakevdp.github.io/PythonDataScienceHandbook/04.04-density-and-contour-plots.html

ndisc=int(np.sqrt(post_params['evidence'].size))

Z=post_params['evidence'].values.reshape((ndisc,ndisc),order='F') #vraie post
Z_koh=post_params['evidence_koh'].values.reshape((ndisc,ndisc),order='F') #post approchée
Z_classical=post_params['evidence_classical'].values.reshape((ndisc,ndisc),order='F')
T1=post_params['t1'].values.reshape((ndisc,ndisc),order='F')[0,:]
T2=post_params['t2'].values.reshape((ndisc,ndisc),order='F')[:,0]
T1,T2=np.meshgrid(T1,T2)

plt.figure(figsize=[16,7])
ax=plt.subplot(121)
cs=plt.contourf(T1,T2,Z)
plt.title('Vraie log-postérieure méthode Opti, nobs = '+str(nobs))
plt.colorbar()
#plt.scatter(x='t1',y='t2',data=samples,marker='+',label='echantillons')
#plt.legend(loc='upper left')
plt.scatter(x='t1',y='t2',data=samples_opti[::4],marker='+')
ax.set_xlabel('t1')
ax.set_ylabel('t2')


ax=plt.subplot(122)
plt.contourf(T1,T2,Z_koh,levels=cs.levels) #on copie les niveaux du plot précédent pour avoir la même échelle de couleurs
plt.title('Log-postérieure koh')
plt.colorbar()
plt.scatter(x='t1',y='t2',data=samples_koh[::4],marker='+')
ax.set_xlabel('t1')
ax.set_ylabel('t2')

if(save):
    plt.savefig(file_name+'_logpost.png')
```

```python
#Contourplot de la postérieure. On peut avoir des couleurs continues avec imshow, voir https://jakevdp.github.io/PythonDataScienceHandbook/04.04-density-and-contour-plots.html

ndisc=int(np.sqrt(post_gp['evidence'].size))

Z=post_gp['evidence'].values.reshape((ndisc,ndisc),order='F') #vraie post
Z_koh=post_gp['evidence_koh'].values.reshape((ndisc,ndisc),order='F') #post approchée
Z_classical=post_gp['evidence_classical'].values.reshape((ndisc,ndisc),order='F')
T1=post_gp['t1'].values.reshape((ndisc,ndisc),order='F')[0,:]
T2=post_gp['t2'].values.reshape((ndisc,ndisc),order='F')[:,0]
T1,T2=np.meshgrid(T1,T2)

plt.figure(figsize=[16,7])
ax=plt.subplot(121)
cs=plt.contourf(T1,T2,Z)
plt.title('GP log-postérieure méthode Opti, nobs = '+str(nobs))
plt.colorbar()
#plt.scatter(x='t1',y='t2',data=samples,marker='+',label='echantillons')
#plt.legend(loc='upper left')
ax.set_xlabel('t1')
ax.set_ylabel('t2')


ax=plt.subplot(122)
plt.contourf(T1,T2,Z_koh,levels=cs.levels) #on copie les niveaux du plot précédent pour avoir la même échelle de couleurs
plt.title('GP Log-postérieure koh')
plt.colorbar()
ax.set_xlabel('t1')
ax.set_ylabel('t2')

if(save):
    plt.savefig(file_name+'_surrogatelogpost.png')
```

```python
#Contourplot de sig_edm.

ndisc=int(np.sqrt(hpars['t1'].size))

EDM=hpars['edm'].values.reshape((ndisc,ndisc),order='F') #vraie post
EXP=hpars['exp'].values.reshape((ndisc,ndisc),order='F') #post approchée
LCOR=hpars['lcor'].values.reshape((ndisc,ndisc),order='F') #post approchée
T1=hpars['t1'].values.reshape((ndisc,ndisc),order='F')[0,:]
T2=hpars['t2'].values.reshape((ndisc,ndisc),order='F')[:,0]
T1,T2=np.meshgrid(T1,T2)

plt.figure(figsize=[14,14])
ax=plt.subplot(221)
cs=plt.contourf(T1,T2,EDM)#,levels=np.linspace(0,5,10)) #niveaux configurés pour ce plot
plt.title('sig_edm optimisé')
plt.colorbar()
ax.set_xlabel('t1')
ax.set_ylabel('t2')


ax=plt.subplot(222)
plt.contourf(T1,T2,EXP) #on copie les niveaux du plot précédent pour avoir la même échelle de couleurs
plt.title('sig_exp optimisé')
plt.colorbar()
ax.set_xlabel('t1')
ax.set_ylabel('t2')

ax=plt.subplot(223)
plt.contourf(T1,T2,LCOR) #on copie les niveaux du plot précédent pour avoir la même échelle de couleurs
plt.title('lcor optimisé')
plt.colorbar()
ax.set_xlabel('t1')
ax.set_ylabel('t2')


if(save):
    plt.savefig(file_name+'_hpars.png')
```

```python
#Affichage des observations de la réalité

def f(x):
    return np.sin(20*x)
#print(predictions)
plt.figure(figsize=[30,15])
ax=plt.subplot(221)
ax.scatter(x='x',y='obs',label='observations',data=observations,marker='+',linewidth=2,color='tab:red')
ax.plot(pred['x'],f(pred['x']),label='truth',color='tab:green')

ax.plot('x','f+zmean',data=pred,label='f+zmean',color='tab:orange')

ax.fill_between('x','f+zlow','f+zsup',data=pred,alpha=0.2,color='tab:orange')
ax.plot('x','f+zlow',data=pred,color='black',linewidth=0.5,label='')
ax.plot('x','f+zsup',data=pred,color='black',linewidth=0.5,label='')


ax.legend(loc='lower left')
plt.title('Predictions méthode Opti, nobs = '+str(nobs))
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid()
ax.set_xlim(0,1)
ax.set_ylim(-3,3)



ax2=plt.subplot(223)
ax2.scatter(x='x',y='obs',label='observations',data=observations,marker='+',linewidth=2,color='tab:red')


ax2.plot(pred['x'],f(pred['x']),label='truth',color='tab:green')

ax2.plot('x','fmean',data=pred,label='fmean',color='tab:blue')
ax2.fill_between('x','flow','fsup',data=pred,alpha=0.2,color='tab:blue')
ax2.plot('x','fsup',data=pred,color='black',linewidth=0.5,label='')
ax2.plot('x','flow',data=pred,color='black',linewidth=0.5,label='')

ax2.plot('x','zmean',data=pred,label='zmean',color='tab:red')
ax2.fill_between('x','zlow','zsup',data=pred,alpha=0.2,color='tab:red')
ax2.plot('x','zsup',data=pred,color='black',linewidth=0.5,label='')
ax2.plot('x','zlow',data=pred,color='black',linewidth=0.5,label='')

ax2.legend(loc='upper left')
ax2.grid()

ax2.set_xlim(0,1)
ax2.set_ylim(-3,3)

plt.title('Décomposition de fz en f et z, méthode Opti')

ax3=plt.subplot(222)
ax3.scatter(x='x',y='obs',label='observations',data=observations,marker='+',linewidth=2,color='tab:red')
ax3.plot(pred_koh['x'],f(pred_koh['x']),label='truth',color='tab:green')

ax3.plot('x','f+zmean',data=pred_koh,label='f+zmean',color='tab:orange')

ax3.fill_between('x','f+zlow','f+zsup',data=pred_koh,alpha=0.2,color='tab:orange')
ax3.plot('x','f+zlow',data=pred_koh,color='black',linewidth=0.5,label='')
ax3.plot('x','f+zsup',data=pred_koh,color='black',linewidth=0.5,label='')


ax3.legend(loc='lower left')
plt.title('Prédictions méthode KOH, nobs = '+str(nobs))
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.grid()
ax3.set_xlim(0,1)
ax3.set_ylim(-3,3)



ax4=plt.subplot(224)
ax4.scatter(x='x',y='obs',label='observations',data=observations,marker='+',linewidth=2,color='tab:red')
ax4.plot(pred_koh['x'],f(pred_koh['x']),label='truth',color='tab:green')
ax4.plot('x','fmean',data=pred_koh,label='fmean',color='tab:blue')
ax4.fill_between('x','flow','fsup',data=pred_koh,alpha=0.2,color='tab:blue')
ax4.plot('x','fsup',data=pred_koh,color='black',linewidth=0.5,label='')
ax4.plot('x','flow',data=pred_koh,color='black',linewidth=0.5,label='')

ax4.plot('x','zmean',data=pred_koh,label='zmean',color='tab:red')
ax4.fill_between('x','zlow','zsup',data=pred_koh,alpha=0.2,color='tab:red')
ax4.plot('x','zsup',data=pred_koh,color='black',linewidth=0.5,label='')
ax4.plot('x','zlow',data=pred_koh,color='black',linewidth=0.5,label='')

ax4.legend(loc='upper left')
ax4.grid()
plt.title('Décomposition de f+z en f et z, méthode KOH')
ax4.set_xlim(0,1)
ax4.set_ylim(-3,3)


if(save):
    plt.savefig(file_name+'_predictions.png')

```
